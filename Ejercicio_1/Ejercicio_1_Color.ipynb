a-Cargar la imagen con OpenCV y mostrar los canales BGR por separado. 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_bgr = cv2.imread('telefe.jpg')
#img_bgr = cv2.imread('messi.jpg')

# Informaci√≥n b√°sica de la imagen
print("Dimensiones de la imagen:", img_bgr.shape)
print("Tipo de datos:", img_bgr.dtype)
print("Tama√±o en memoria (bytes):", img_bgr.nbytes)

# Visualizaci√≥n de la imagen completa
plt.imshow(img_bgr)
```

Canal Azul (B)

```python
canal_B = img_bgr[:,:,0]
print(f"M√≠nimo: {canal_B.min()}, M√°ximo: {canal_B.max()}, Promedio: {canal_B.mean():.2f}")
```

```python
plt.imshow(canal_B, cmap='gray')
```

```python
plt.imshow(canal_B, cmap='Blues')
```

Canal Verde (G)

```python
canal_G = img_bgr[:,:,1]
print(f"M√≠nimo: {canal_G.min()}, M√°ximo: {canal_G.max()}, Promedio: {canal_G.mean():.2f}")
#plt.imshow(canal_G, cmap='gray')
```

```python
plt.imshow(canal_B, cmap='gray')
```

```python
plt.imshow(canal_B, cmap='Greens')
```

Canal Rojo (R)

```python
canal_R = img_bgr[:,:,2]
print(f"M√≠nimo: {canal_R.min()}, M√°ximo: {canal_R.max()}, Promedio: {canal_R.mean():.2f}")
#plt.imshow(canal_R, cmap='gray')
```

```python
plt.imshow(canal_B, cmap='gray')
```

```python
plt.imshow(canal_B, cmap='Reds')
```

b-Identificar cu√°l de los tres canales tiene mayor informaci√≥n bas√°ndose en los valores promedio. 

```python
promedio_b = np.mean(canal_B)
promedio_g = np.mean(canal_G)
promedio_r = np.mean(canal_R)
print(f'Promedio Azul (B): {promedio_b:.2f}')
print(f'Promedio Verde (G): {promedio_g:.2f}')
print(f'Promedio Rojo (R): {promedio_r:.2f}')
```

El canal con mayor informaci√≥n promedio es el **Rojo**

c-Convertir la imagen de BGR a RGB y explicar por qu√© los colores se ven diferentes antes y despu√©s de la conversi√≥n.

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.title('Imagen RGB')
plt.imshow(img_rgb)

plt.show()
```

La diferencia radica en el √≥rden en el que manejan el √≥rden de los canales. OpenCV almacena primero el Azul, despu√©s el Verde y finalmente el Rojo, mientras que RGB (primero Rojo, despues Verde y √∫ltimo el Azul) es el formato al que m√°s estamos acostumbrados a visualizar.
